# **EXPERIMENT 1**

```python
import pandas as pd
df= ('/content/drive/MyDrive/winequality-red[1].csv')
red= pd.read_csv( df, delimiter=';')
df1=('/content/drive/MyDrive/winequality-white[1].csv')
white= pd.read_csv(df1,delimiter=';')
red['type']='red'
white['type']='white'

#print(red)
#print(white)
mergedf=pd.merge(red,white,how="outer")
mergedf
```

# **EXPERIMENT 2**

**Handling misssing values**

```python
print(mergedf.isnull().sum())
```

**Removing duplicates**

```python
print(mergedf.drop_duplicates().sum())
```

**Handling outliers**

```python
q1=mergedf['alcohol'].quantile(0.25)
q3=mergedf['alcohol'].quantile(0.75)
iqr=q3-q1
cleandf=mergedf[(mergedf['alcohol'] > q1-1.5*iqr) & (mergedf['alcohol'] < q3+1.5*iqr)]
cleandf
```

# **EXPERIMENT 3**

**LABEL ENCODING**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
mergedf["type"] = le.fit_transform(mergedf["type"])
mergedf
```

**ONE-HOT ENCODING**

```python
import pandas as pd

one_hot = pd.get_dummies(mergedf['type'])

one_hot
```

**ORDINAL ENCODING**

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

quality_levels = sorted(mergedf["quality"].unique())

encoder = OrdinalEncoder(categories=[quality_levels])

mergedf["quality_encoded"] = encoder.fit_transform(mergedf[["quality"]])
mergedf
```
