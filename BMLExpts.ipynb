## Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df[' State-gov'] = label_encoder.fit_transform(df[' State-gov'])
df[' <=50K']= label_encoder.fit_transform(df[' <=50K'])
df[' Male']= label_encoder.fit_transform(df[' Male'])
df[' Bachelors']= label_encoder.fit_transform(df[' Bachelors'])
df[' Not-in-family']= label_encoder.fit_transform(df[' Not-in-family'])
df[' White']= label_encoder.fit_transform(df[' White'])
df[' Adm-clerical']= label_encoder.fit_transform(df[' Adm-clerical'])
df[' Never-married']= label_encoder.fit_transform(df[' Never-married'])
df[' United-States']= label_encoder.fit_transform(df[' United-States'])
df
```

## identifying variables & splitting data

```python
from sklearn.model_selection import train_test_split
# x= df[['39','77516' 'State-gov',' Male',' Bachelors',' Not-in-family','13','Never-married','Adm-clerical','White','2174','United-states']]
x= df.drop(columns=[' <=50K'])
y= df[[' <=50K']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()
model.fit(x_train, y_train)
score_train = model.score(x_train, y_train)
score_test = model.score(x_test, y_test)
y_pred = model.predict (x_test)
```

```python
score_train
```

```python
score_test
```

## Logistic regression

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
log_reg.score(x_test,y_test)
```

# **Expt 4** : *Naive bayes*

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
y_pred_nb = nb.predict(x_test)

print("Naïve Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
```

# **Expt 7** : *KNN*

```python
from sklearn.neighbors import KNeighborsClassifier
k = 5
knn = KNeighborsClassifier(n_neighbors=k)


knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

# **Expt 8** : *k-means clustering*

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=3, random_state=42)  # Set k=3, as there are 3 types of flowers
kmeans.fit(df)
centroids = kmeans.cluster_centers_

# Use PCA to reduce the dataset to 2 dimensions for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)

# Plot the data points with their predicted clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=kmeans.labels_, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], color='red', marker='X', s=200, label='Centroids')

# Adding labels and title
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```
